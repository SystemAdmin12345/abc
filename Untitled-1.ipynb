```python
# Functional Defenitions
def add(a, b):
    ans = a + b
    return ans

def sub(a, b):
    ans = a - b
    return ans

def mul(a, b):
    ans = a * b
    return ans

def div(a, b):
    ans = a / b
    return ans

def mod(a, b):
    ans = a % b
    return ans

def pow(a, b):
    ans = a ** b
    return ans

def lshift(a, b):
    ans = a << b
    return ans

def rshift(a, b):
    ans = a >> b
    return ans

def and_(a, b):
    ans = a & b
    return ans

def or_(a, b):
    ans = a | b
    return ans

def xor(a, b):
    ans = a ^ b
    return ans

def neg(a):
    ans = -a
    return ans

def not_(a):
    ans = ~a
    return ans

def is_(a, b):
    ans = a is b
    return ans

def is_not(a, b):
    ans = a is not b
    return ans

def in_(a, b):
    ans = a in b
    return ans

def not_in(a, b):
    ans = a not in b
    return ans

def is_true(a):
    ans = a is True
    return ans

def is_not_true(a):
    ans = a is not True
    return ans

def is_false(a):
    ans = a is False
    return ans

def is_not_false(a):
    ans = a is not False
    return ans

def is_none(a):
    ans = a is None
    return ans

def is_not_none(a):
    ans = a is not None
    return ans

def is_int(a):
    ans = isinstance(a, int)
    return ans

def is_not_int(a):
    ans = not is_int(a)
    return ans

def is_float(a):
    ans = isinstance(a, float)
    return ans

def is_not_float(a):
    ans = not is_float(a)
    return ans

def is_str(a):
    ans = isinstance(a, str)
    return ans

def is_not_str(a):
    ans = not is_str(a)
    return ans

def is_list(a):
    ans = isinstance(a, list)
    return ans

def is_not_list(a):
    ans = not is_list(a)
    return ans

def is_tuple(a):
    ans = isinstance(a, tuple)
    return ans

def is_not_tuple(a):
    ans = not is_tuple(a)
    return ans

def is_dict(a):
    ans = isinstance(a, dict)
    return ans

def is_not_dict(a):
    ans = not is_dict(a)
    return ans

def is_set(a):
    ans = isinstance(a, set)
    return ans

def is_not_set(a):
    ans = not is_set(a)
    return ans

def is_bool(a):
    ans = isinstance(a, bool)
    return ans

def is_not_bool(a):
    ans = not is_bool(a)
    return ans

def is_none_type(a):
    ans = isinstance(a, type(None))
    return ans

def is_not_none_type(a):
    ans = not is_none_type(a)
    return ans

def is_int_type(a):
    ans = isinstance(a, int)
    return ans

def is_not_int_type(a):
    ans = not is_int_type(a)
    return ans
```
